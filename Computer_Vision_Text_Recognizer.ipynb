```python
# best
import cv2
import numpy as np
import pytesseract
from google.colab import files
import re

# Load image
uploaded = files.upload()
image = cv2.imread(list(uploaded.keys())[0])

# Resize image to a higher resolution
image = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

# Convert image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Enhance image quality
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
gray = clahe.apply(gray)
gray = cv2.medianBlur(gray, 3)

# Perform Otsu's threshold
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

# Define kernel for morphological operations
kernel = np.ones((3,3), np.uint8)
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

# Set Tesseract configuration
config = ('-l eng --oem 1 --psm 3')

# Extract text from image
text = pytesseract.image_to_string(closing, config=config)

# Post-processing
text = re.sub(r'\n+', '\n', text) # Remove extra newlines
text = text.strip() # Remove leading/trailing whitespaces

# Print the extracted text
print(text)
```

```python
# image 2
import cv2
import numpy as np
import pytesseract
from google.colab import files

# Load image
uploaded = files.upload()
image = cv2.imread(list(uploaded.keys())[0])

# Convert image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Perform Gaussian blur to remove noise
blur = cv2.GaussianBlur(gray, (5,5), 0)

# Perform thresholding to segment text
thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

# Perform erosion and dilation to remove small noise
kernel = np.ones((3,3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=1)

# Perform denoising using fastNlMeansDenoising
denoised = cv2.fastNlMeansDenoising(opening, None, 10, 7, 21)

# Set Tesseract configuration
config = ('-l eng --oem 1 --psm 6')

text = pytesseract.image_to_string(denoised, config=config)
print(text)
```
